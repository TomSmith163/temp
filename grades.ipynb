```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
data = pd.read_excel(r"C:\大三上\前四学期专业排名（重修前，无姓名）\人工智能.xlsx",header=[0,1,2,3,4])[:-1]
data.columns = ["名次","学号","2020年秋季","2021年春季","2021年秋季","2022年春季","2022年秋季","2023年春季","总平均加权成绩","总学分","班级"]
data
rank = data[["2020年秋季","2021年春季","2021年秋季","2022年春季","2022年秋季","2023年春季","总平均加权成绩"]].rank(axis=0,method="min",ascending=False)
rank.columns = ["2020年秋季排名","2021年春季排名","2021年秋季排名","2022年春季排名","2022年秋季排名","2023年春季排名","总平均加权成绩排名"]
rank
data_p = pd.concat([data,rank],axis=1)
data_p
```

```python
class2101 = data_p[data_p["班级"] == "人工智能202101班"].copy()
class2101
```

```python
#学年内进步
class2101["学年内进步"] = class2101["2023年春季"] - class2101["2022年秋季"]
class2101["上学期同比进步"] = class2101["2023年春季"] - class2101["2022年春季"]

class2101["大二上学期环比进步"] = class2101["2022年春季排名"] - class2101["2022年秋季排名"]
class2101["大二下学期环比进步"] = class2101["2022年秋季排名"] - class2101["2023年春季排名"]
class2101
```

```python
plt.plot(class2101["学年内进步"],class2101["上学期同比进步"],"o")
plt.xlim(-10,10)
plt.ylim(-10,10)
xlabel = plt.xlabel("last academic year progress")
ylabel = plt.ylabel("last semester progress")
for x,y,z in zip(class2101["学年内进步"],class2101["上学期同比进步"],class2101["学号"]):
    label = "{}".format(z[-5:])
    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
plt.savefig(r"C:\大三上\pyjects\out\pic1.png",dpi=300)
```

```python
plt.plot(class2101["大二上学期环比进步"],class2101["大二下学期环比进步"],"o")
plt.xlim(-5,)
plt.ylim(-5,)
plt.xlabel("freshman progress")
plt.ylabel("sophomore progress")
for x,y,z in zip(class2101["大二上学期环比进步"],class2101["大二下学期环比进步"],class2101["学号"]):
    label = "{}".format(z[-5:])
    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
plt.savefig(r"C:\大三上\pyjects\out\pic2.png",dpi=300)
```
